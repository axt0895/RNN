```python
import pandas as pd
from pathlib import Path
```

```python
path = Path('/Users/anilthapa/Downloads/CTA_-_Ridership_-_Daily_Boarding_Totals_20241114.csv')
df = pd.read_csv(path, parse_dates=['service_date'])
```

```python
df.columns = ['date', 'day_type', 'bus', 'rail', 'total']
```

```python
df.head()
```

```python
df = df.sort_values('date').set_index('date')
df
```

```python
df.drop('total', axis = 1, inplace = True)
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.shape
```

```python
df.head(10)
```

```python
import matplotlib.pyplot as plt
```

```python
df['2019-01': '2019-03']['bus'].plot(grid = True, figsize = (9, 6), marker = '.')
plt.show()
```

```python
diff_7 = df[['bus', 'rail']].diff(7)['2019-03':'2019-05']
fig, ax = plt.subplots(2, 1, sharex=True, figsize = (8, 5))

df.plot(ax = ax[0], legend = False, marker = '.')
df.shift(7).plot(ax = ax[0], legend = False, linestyle = ':')
diff_7.plot(ax = ax[1], legend = False, marker = '.')
```
